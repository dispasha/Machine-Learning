```python
import numpy as np
from sklearn.metrics import roc_auc_score
```

```python
data = np.genfromtxt('data-logistic.csv', delimiter=',')
```

```python
y = data[:,0]
y
```

```python
X = data[:,1:3]
X
```

```python
def grad_des(X,y, C=0, k=0.1):
    X = np.transpose(X)
    w = np.zeros(2)
    l = y.shape[0]
    for tries in range(10000):
        w_new = np.array(\
        [w[0]+k*np.sum(y*X[0]*(1-1/(1+np.exp(-y*(w[0]*X[0]+w[1]*X[1])))))/l-k*C*w[0],\
        w[1]+k*np.sum(y*X[1]*(1-1/(1+np.exp(-y*(w[0]*X[0]+w[1]*X[1])))))/l-k*C*w[1]])
        if np.linalg.norm(w_new-w) < 1e-5:
            print('good')
            return 1/(1+np.exp(-w_new[0]*X[0]-w_new[1]*X[1]))
        w=w_new
    return w
```

```python
ocenki = grad_des(X,y)
```
