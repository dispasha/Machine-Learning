```python
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
```

```python
X = np.array([[1, 2], [3, 4], [5, 6]])
y = np.array([0, 1, 0])
clf = DecisionTreeClassifier()
clf.fit(X, y)
tree.plot_tree(clf)
```

```python
importances = clf.feature_importances_
```

```python
importances
```

```python
import pandas as pd
titanic_data = pd.read_csv("titanic.csv")
```

```python
titanic_data
```

```python
X = pd.concat([titanic_data.Pclass, titanic_data.Fare, titanic_data.Age, titanic_data.Sex=='male', titanic_data.Survived], axis=1).dropna()
Y = X.Survived
X = X.drop(labels = "Survived", axis=1)
X
```

```python
clf = DecisionTreeClassifier(random_state = 241)
clf.fit(X, Y)
importances = clf.feature_importances_
importances
```
