```python
import pandas as pd
import numpy as np
import sklearn
from sklearn.metrics import precision_recall_curve, roc_auc_score, accuracy_score, precision_score, recall_score, f1_score
```

```python
cl = pd.read_csv('classification.csv')
cl = cl.replace({0:False, 1:True})
```

```python
y_t = cl['true']
y_p = cl['pred']
```

```python
TP = (y_t & y_p).value_counts()[True]
TN = (np.invert(y_t) & np.invert(y_p)).value_counts()[True]
FN = (y_t & np.invert(y_p)).value_counts()[True]
FP = (np.invert(y_t) & y_p).value_counts()[True]
```

```python
print(TP, FP, FN, TN)
```

```python
print(accuracy_score(y_t, y_p), precision_score(y_t, y_p), recall_score(y_t, y_p), f1_score(y_t, y_p))
```

```python
clfs = pd.read_csv('scores.csv')
```

```python
for i in clfs.drop('true', axis=1).columns:
    print(i, roc_auc_score(clfs['true'], clfs[i]))
```

```python
for i in clfs.drop('true', axis=1).columns:
    pr_rec = precision_recall_curve(clfs['true'], clfs[i])[0:2]
    pr_rec = pd.DataFrame(np.transpose(np.array([pr_rec[0], pr_rec[1]])))
    print(i, np.max(pr_rec[pr_rec[1]>0.7][0]))
```
