```python
import sklearn as sk
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsRegressor
from sklearn import datasets
```

```python
boston = datasets.load_boston()
```

```python
X = sk.preprocessing.scale(boston['data'])
y = boston['target']
```

```python
X
```

```python
p_values = np.linspace(1, 10, 200)
```

```python
from sklearn.model_selection import KFold
frag = KFold(5, shuffle=True, random_state=42)
def find_p(X, y, p_values):
    scores = []
    for p in p_values:
        reg = KNeighborsRegressor(n_neighbors=5, weights='distance', p=p)
        score = np.mean(sk.model_selection.cross_val_score(reg, X, y, cv=frag, scoring='neg_mean_squared_error'))
        scores.append(score)
    return pd.DataFrame(scores, p_values).mean(axis=1).sort_values(ascending=False)    
```

```python
find_p(X,y,p_values)
```
