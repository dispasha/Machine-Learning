```python
import pandas as pd
import sklearn
from sklearn.neighbors import KNeighborsClassifier as KN
from sklearn.model_selection import KFold
import numpy as np
```

```python
wine_data = pd.read_csv('wine.data')
```

```python
wine_data
```

```python
frag = KFold(5, shuffle=True, random_state=42)
```

```python
X = sklearn.preprocessing.scale(wine_data.drop('Class', axis=1))
_X = wine_data.drop('Class', axis=1).to_numpy()
_X.shape
```

```python
Y = wine_data.Class.to_numpy()
Y
```

```python
def accuracy(kf, X, y):
    scores = []
    scores_range = range(1,51)
    for k in scores_range:
        neigh = KN(n_neighbors=k)
        score = np.mean(sklearn.model_selection.cross_val_score(neigh, X, y, cv=kf, scoring='accuracy'))
        scores.append(score)
    return pd.DataFrame(scores, scores_range).mean(axis=1).sort_values(ascending=False)
accuracy(frag, _X, Y)
```

```python
accuracy(frag, X, Y)
```
